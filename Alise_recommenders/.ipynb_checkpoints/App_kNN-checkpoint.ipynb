```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

from sklearn.utils import shuffle
from sklearn.utils import check_random_state
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
from sklearn.metrics import pairwise_distances                 
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation
from sklearn.datasets import make_blobs
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.neighbors import NearestNeighbors, DistanceMetric

from scipy.sparse import csr_matrix 

import pandas as pd
import numpy as np
import sys      
import os
%matplotlib inline
```

# Cleanup app dataset to get app that are functional and have been run

```python
# import text documents from app dataset
app_data=pd.read_csv('../../preprocess_data/preprocessed/app_info_dataset_2020_05_01_cleaned.csv')
app_data = app_data.set_index(['app_id'])
app_data.head()
```

```python
# select apps that were run at least once
has_run =  app_data.dropna(subset=['job_count'])
has_run.head()
```

```python
# remove apps that have been deleted
not_deleted = has_run['deleted']==False
app_selection = has_run[not_deleted]
# remove apps that are disabled
not_disabled = app_selection['disabled']==False
app_selection = app_selection[not_disabled]
# remove apps that have been run less than 10 times
run_enough = app_selection['job_count']>10
app_selection = app_selection[run_enough]
```

```python
# select relevant columns
dataset_desc= app_selection[['name', 'description', 'job_count']]
dataset_desc=dataset_desc.drop_duplicates()
dataset_desc.head()
```

# Kmeans on the app description
## TI-IDF model of the apps descriptions

```python
# extract words from description
dataset_desc['description']=dataset_desc['description'].apply(lambda x: np.str_(x))
# get unique apps descriptions
#app_desc=app_desc.drop_duplicates()
```

```python
# get word counts from descriptions
#Define the TFIDF vectorizer that will be used to process the data
tfidf_vectorizer = TfidfVectorizer()
#Apply this vectorizer to the full dataset to create normalized vectors
tf_idf = tfidf_vectorizer.fit_transform(dataset_desc['description'])
```

## Choose number of K

```python
def plot_k_vs_heterogeneity(k_values, heterogeneity_values):
    plt.figure(figsize=(7,4))
    plt.plot(k_values, heterogeneity_values, linewidth=4)
    plt.xlabel('K')
    plt.ylabel('Inertia')
    plt.title('K vs. Inertia')
    plt.rcParams.update({'font.size': 16})
    plt.tight_layout()
```

```python
heterogeneity_values = []
k_list = [2, 10, 25, 50, 100, 300, 500]
```

```python
for n_clusters in k_list:
    km = KMeans(n_clusters=n_clusters, init='k-means++').fit(tf_idf)
    heterogeneity_values.append(km.inertia_)
```

```python
plot_k_vs_heterogeneity(k_list, heterogeneity_values)
```

## K-means

```python
# compute kmeans using kmeans++ initialization
kmeans = KMeans(n_clusters=100, init='k-means++').fit(tf_idf)  
# predict the closest cluster for each point
labels= kmeans.predict(tf_idf)
```

```python
# retrieve other pages from a cluster
def get_cluster(cluster_nb, labels, dataset):
    curr_cluster=[]
    name_app=[]
    for index in range(len(labels)):
        if labels[index]==cluster_nb:
            curr_cluster.append(dataset.index.values[index,])
            name_app.append(dataset.iloc[[index,],[0]].to_string(header=False, index=False))
            
    result = pd.DataFrame({'app_id': curr_cluster,
                           'name': name_app})
    return result
        
```

```python
# Find the cluster in which 7d94d294-9876-11e4-a4da-d7ead82e9f22 is
pos = dataset_desc.index.get_loc("7d94d294-9876-11e4-a4da-d7ead82e9f22")
labels[pos]
```

```python
cluster= get_cluster(labels[pos], labels, dataset_desc)
cluster
```

# K-NN search on the TF-IDF
## Euclidean distance of the documents

```python
nbrs = NearestNeighbors(n_neighbors=10, metric='cosine').fit(tf_idf)
distances, indices = nbrs.kneighbors()

def get_closest_neighs(app_id):
    row = dataset_desc.index.get_loc(app_id)
    result = pd.DataFrame({'app_id':dataset_desc.iloc[indices[row],].reset_index()['app_id'],
                           'distance': distances[row].flatten(),
                          'app name': dataset_desc.iloc[indices[row],].reset_index()['name']})
    return result
```

```python
get_closest_neighs("7d94d294-9876-11e4-a4da-d7ead82e9f22")
```
